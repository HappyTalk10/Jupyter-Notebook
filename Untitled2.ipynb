```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 日本語フォントを検索し、設定する
font_path = '/usr/share/fonts/opentype/noto/NotoSansCJK-Regular.ttc'  # 多くのCodespaces環境で利用可能なフォント
font_prop = fm.FontProperties(fname=font_path, size=14)
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Noto Sans CJK JP']  # 警告を解消するために明示的に設定
plt.rcParams['axes.unicode_minus'] = False # マイナス記号の表示を正しくするため

# グラフ描画コード
x = [1, 2, 3, 4, 5]
y = [i**2 for i in x]

plt.plot(x, y)
plt.title("y = x^2", fontproperties=font_prop)
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
import matplotlib as mpl
import os

# フォントキャッシュのパスを取得
cache_dir = mpl.get_cachedir()
# キャッシュディレクトリ内のフォントキャッシュを削除
if os.path.exists(cache_dir):
    for file in os.listdir(cache_dir):
        if file.startswith('fontlist-'):
            os.remove(os.path.join(cache_dir, file))
            print("Matplotlibのフォントキャッシュを削除しました。")
```
